# Data Cleaning

```python
data_copy=tele_data.copy()
```

Total Charges should be a numeric values

```python
data_copy.TotalCharges=pd.to_numeric(data_copy.TotalCharges,errors='coerce')
data_copy.isnull().sum()
```

Here 11 values are null

```python
data_copy.loc[data_copy['TotalCharges'].isnull()==True]
```

```python
#Removing missing values
data_copy.dropna(how='any',inplace=True)
```

Divide customer into bins based on tenure

```python
print(data_copy['tenure'].max())
```

```python
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
data_copy['tenure_group']=pd.cut(data_copy.tenure,range(1,80,12),right=False,labels=labels)
```

```python
data_copy['tenure_group'].value_counts()
```

```python
data_copy.drop(columns=['customerID','tenure'],axis=1,inplace=True)
data_copy.head()
```

# Data Exploration

```python
for i, predictor in enumerate(data_copy.drop(columns=['MonthlyCharges','TotalCharges','Churn'])):
    plt.figure(i)
    sns.countplot(data=data_copy,x=predictor,hue='Churn')
```

```python
#Convert Churn data from Yes or No to Binary Data

data_copy['Churn']=np.where(data_copy.Churn=='Yes',1,0)
```

```python
data_copy.head()
```

```python
data_copy_dummies=pd.get_dummies(data_copy)
data_copy_dummies.head()
```

Relation between

```python
sns.lmplot(data=data_copy_dummies,x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Churn by Monthly Charges and Total charges

```python
mth=sns.kdeplot(data_copy_dummies.MonthlyCharges[(data_copy_dummies["Churn"]==0)],color="Red",shade=True)
mth=sns.kdeplot(data_copy_dummies.MonthlyCharges[(data_copy_dummies["Churn"]==1)], ax=mth,color="Blue",shade=True)
mth.legend(["No Churn","Churn"],loc='upper right')
mth.set_ylabel('Density')
mth.set_xlabel('Monthly Charges')
mth.set_title('Monthly Charges by Churn')
```

Churning is significantly reduced when monthly charges are less and churning increases as monthly charges increases

```python
tot=sns.kdeplot(data_copy_dummies.TotalCharges[(data_copy_dummies["Churn"]==0)],color="Red",shade=True)
tot=sns.kdeplot(data_copy_dummies.TotalCharges[(data_copy_dummies["Churn"]==1)],ax=tot,color="Blue",shade=True)
tot.legend(["No Churn","Churn"],loc="upper right")
tot.set_ylabel('Density')
tot.set_xlabel('Total Charges')
tot.set_title('Total Charges by Churn')
```

```python
plt.figure(figsize=(20,8))
data_copy_dummies.corr()["Churn"].sort_values(ascending=False).plot(kind='bar')
```

1. If the contract length is month-to-month then we can see the maximum number of churns where as in case of yearly contract churning is minimum

# Bivariate Analysis

```python
new_df0=data_copy.loc[data_copy['Churn']==0]
new_df1=data_copy.loc[data_copy['Churn']==1]
```

```python
def uniplot(df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"]=20
    plt.rcParams["axes.titlesize"]=22
    plt.rcParams["axes.titlepad"]=30
    
    temp=pd.Series(data=hue)
    fig, ax=plt.subplots()
    width=len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax=sns.countplot(data=df,x=col,order=df[col].value_counts().index,hue=hue,palette='bright')
    plt.show()
```

```python
uniplot(new_df1,col='Partner',title="Distribution of Gender for Churned Customer",hue='gender')
```

When there is no partner there are 50-50 chance that both male and female will churn

When there is a partner men are most likely to churn

```python
uniplot(new_df1,col="PaymentMethod",title="Distribution of Payment method for Churned Customers",hue='gender')
```

```python
uniplot(new_df1,col="Contract",title="Monthly Contract",hue='gender')
```

```python
uniplot(new_df1,col='TechSupport',title="Churners on the basis of Tech Support",hue='gender')
```

```python
uniplot(new_df1,col="tenure_group",title="Churners on The Basis of Tenure group",hue='gender')
```

```python
uniplot(new_df1,col='SeniorCitizen',title="Churners on the basis of senior citizen",hue='gender')
```
